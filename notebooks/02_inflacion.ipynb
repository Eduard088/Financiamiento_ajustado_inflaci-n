## Creamos los Datos de Financiamiento:

```python
financiamiento_original = {
    1997: 2446723214.77, 1998: 1046566954.67, 1999: 1312364397.62, 2000: 3064092232.97,
    2001: 2250685592.84, 2002: 2439871186.19, 2003: 4942596068.79, 2004: 1854015953.66,
    2005: 2013252441.28, 2006: 4171096908.42, 2007: 2704100762.62, 2008: 2690311483.58,
    2009: 3631639027.92, 2010: 2997358834.02, 2011: 3212932808.88, 2012: 5142514887.61,
    2013: 3670843752.88, 2014: 4002018895.98, 2015: 5122198499.05, 2016: 4031121882.70,
    2017: 4138727087.00, 2018: 6745554160.00, 2019: 4965828351.00, 2020: 5239001651.00,
    2021: 6871356602.00, 2022: 5821851702.00, 2023: 6233510798.00
}
```

## Creamos Listas para Almacenar los Datos:

```python
anios = []
tasas_inflacion = []
financiamiento_publico_original = []
financiamiento_con_inflacion = []
```

## Iteramos para Generar los Resultados:

```python
for año in range(1998, 2024):  # Comienza en 1998 para comparar con 1997 y avanza hasta 2023
    inflacion_anual = ((promedio_inpc_anual[año] - promedio_inpc_anual[año - 1]) / promedio_inpc_anual[año - 1]) * 100
    financiamiento_estimado = financiamiento_original[año - 1] * (1 + inflacion_anual / 100)

    # Almacenar datos en las listas
    anios.append(año)
    tasas_inflacion.append(inflacion_anual)
    financiamiento_publico_original.append(financiamiento_original[año])
    financiamiento_con_inflacion.append(financiamiento_estimado)
```

## Creamos el Dataframe:

```python
df_comparativo = pd.DataFrame({
    "Año": anios,
    "Inflación (%)": tasas_inflacion,
    "Financiamiento Público Original": financiamiento_publico_original,
    "Financiamiento con Inflación": financiamiento_con_inflacion
})
```
